```python
import pickle
import re
import matplotlib.pyplot as plt

```

```python
with open('lyrics-updated.pkl', 'rb') as f:
    lyrics = pickle.load(f)

with open('not_found.pkl', 'rb') as f1:
    not_found = pickle.load(f1)

# lengths of the songs
# min: 5, max: 2114

# min_song_length = float('inf')
# max_song_length = float('-inf')

# for key, value in lyrics.items():
#     min_song_length = min(min_song_length, len(re.split(r'[ \n]+', value)))
#     max_song_length = max(max_song_length, len(re.split(r'[ \n]+', value)))

# print('min:', min_song_length)
# print('max:', max_song_length)

doc_lengths = {
    '0-200': 0, '200-400': 0, '400-600': 0, '600-800': 0, 
    '800-1000': 0, '1000-1200': 0, '1200-1400': 0, 
    '1400-1600': 0, '1600-1800': 0, '1800-2000': 0, '2000+': 0
}

# Loop through lyrics to calculate document lengths
for key, value in lyrics.items():
    curr_length = len(re.split(r'[ \n]+', value))
    
    if curr_length <= 200:
        doc_lengths['0-200'] += 1
    
    elif curr_length <= 400:
        doc_lengths['200-400'] += 1
    
    elif curr_length <= 600:
        doc_lengths['400-600'] += 1
    
    elif curr_length <= 800:
        doc_lengths['600-800'] += 1
    
    elif curr_length <= 1000:
        doc_lengths['800-1000'] += 1
    
    elif curr_length <= 1200:
        doc_lengths['1000-1200'] += 1
    
    elif curr_length <= 1400:
        doc_lengths['1200-1400'] += 1
    
    elif curr_length <= 1600:
        doc_lengths['1400-1600'] += 1
    
    elif curr_length <= 1800:
        doc_lengths['1600-1800'] += 1
    
    elif curr_length <= 2000:
        doc_lengths['1800-2000'] += 1
    
    else:
        doc_lengths['2000+'] += 1

print(doc_lengths)

```

```python
buckets = list(doc_lengths.keys())
values = list(doc_lengths.values())

# Create a bar chart
plt.bar(buckets, values)

# Add labels and title
plt.xlabel('Song Length (number of words)')
plt.ylabel('Number of Songs')
plt.title('Song Lengths')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
for i, value in enumerate(values):
    plt.text(i, value + 5, str(value), ha='center')

# Show the plot
plt.tight_layout()
plt.show()
```
