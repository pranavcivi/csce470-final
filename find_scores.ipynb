```python
import pandas as pd
from sklearn.cluster import KMeans
# %pip install openpyxl
```

```python
df = pd.read_excel('lyrics.xlsx', sheet_name='muse_v3')

valence_low = 0.235
arousal_low = 0.108113208
dominance_low = 0.229230769

valence_high = 8.475
arousal_high = 7.27
dominance_high = 7.44

valence_range = valence_high - valence_low
arousal_range = arousal_high - arousal_low
dominance_range = dominance_high - dominance_low

scores = {}
counts = {}

for index, row in df.iterrows():
    seeds = row['seeds'][1:-1].split(',')
    for seed in seeds:
        seed = seed.strip()
        if seed not in scores:
            scores[seed] = {}
            scores[seed]['valence'] = row['valence_tags']
            scores[seed]['arousal'] = row['arousal_tags']
            scores[seed]['dominance'] = row['dominance_tags']
            counts[seed] = 1
        else:
            scores[seed]['valence'] += row['valence_tags']
            scores[seed]['arousal'] += row['arousal_tags']
            scores[seed]['dominance'] += row['dominance_tags']
            counts[seed] += 1
            
for key, value in scores.items():
    scores[key]['valence'] /= counts[key]
    scores[key]['arousal'] /= counts[key]
    scores[key]['dominance'] /= counts[key]





# cutoff_mid_v = (valence_range * (1/2)) + valence_low

# cutoff_mid_a = (arousal_range * (1/2)) + arousal_low

# cutoff_mid_d = (dominance_range * (1/2)) + dominance_low
# this should give 27 categories

```

```python
# find the cutoffs for the ranges
cutoff_low_v = (valence_range * (1/3)) + valence_low
cutoff_mid_v = (valence_range * (2/3)) + valence_low

cutoff_low_a = (arousal_range * (1/3)) + arousal_low
cutoff_mid_a = (arousal_range * (2/3)) + arousal_low

cutoff_low_d = (dominance_range * (1/3)) + dominance_low
cutoff_mid_d = (dominance_range * (2/3)) + dominance_low

categories = {}
for i in range(27):
    categories[f'Group{i+1}'] = set()

for emotion, value in scores.items():
    curr_v = value['valence']
    curr_a = value['arousal']
    curr_d = value['dominance']

    if curr_v <= cutoff_low_v:
        if curr_a <= cutoff_low_a:
            if curr_d <= cutoff_low_d:
                categories['Group1'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group2'].add(emotion)
            else:
                categories['Group3'].add(emotion)
        elif curr_a <= cutoff_mid_a:
            if curr_d <= cutoff_low_d:
                categories['Group4'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group5'].add(emotion)
            else:
                categories['Group6'].add(emotion)
        else:
            if curr_d <= cutoff_low_d:
                categories['Group7'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group8'].add(emotion)
            else:
                categories['Group9'].add(emotion)

    elif curr_v <= cutoff_mid_v:
        if curr_a <= cutoff_low_a:
            if curr_d <= cutoff_low_d:
                categories['Group10'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group11'].add(emotion)
            else:
                categories['Group12'].add(emotion)
        elif curr_a <= cutoff_mid_a:
            if curr_d <= cutoff_low_d:
                categories['Group13'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group14'].add(emotion)
            else:
                categories['Group15'].add(emotion)
        else:
            if curr_d <= cutoff_low_d:
                categories['Group16'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group17'].add(emotion)
            else:
                categories['Group18'].add(emotion)

    else:
        if curr_a <= cutoff_low_a:
            if curr_d <= cutoff_low_d:
                categories['Group19'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group20'].add(emotion)
            else:
                categories['Group21'].add(emotion)
        elif curr_a <= cutoff_mid_a:
            if curr_d <= cutoff_low_d:
                categories['Group22'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group23'].add(emotion)
            else:
                categories['Group24'].add(emotion)
        else:
            if curr_d <= cutoff_low_d:
                categories['Group25'].add(emotion)
            elif curr_d <= cutoff_mid_d:
                categories['Group26'].add(emotion)
            else:
                categories['Group27'].add(emotion)


categories
```

```python
import pickle
classifications = {}
classifications['cynical'] = categories['Group5']
classifications['wealth'] = categories['Group11']
classifications['cynical'] = categories['Group5']
classifications['dark'] = categories['Group14']
classifications['dynamic'] = categories['Group15']
classifications['anger'] = categories['Group17']
classifications['fierce'] = categories['Group18']
classifications['uplifting'] = categories['Group24']
classifications['energetic'] = categories['Group27']

print(classifications)

with open('classifications.pkl', 'wb') as f:
    pickle.dump(classifications, f)
```
